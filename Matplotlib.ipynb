```python
x = np.array([70,75,80,85,90,95,100,105,110,115,120])
y=np.array([240,250,260,270,280,290,300,310,320,330,340])

font1={ 'family':'serif' , 'color': 'blue' , 'size':20}
font2={ 'family':'serif' , 'color': 'darkred' , 'size':15}

plt.title("Fitness watch Data" , fontdict=font1)
plt.xlabel("Average Pulse", fontdict=font2)
plt.ylabel("Calorie Burnage", fontdict=font2)

plt.plot(x,y ,marker='o')
plt.show()
```

```python
###Subplot
## plt.subplot( rows, columns, plot number)

##plot no 1
x= np.array([1,2,3,4])
y= np.array([3,6,8,9])

plt.subplot(2,4,1)
plt.plot(x,y)
plt.title("plot no 1")
plt.xlabel("x axis")
plt.ylabel("y axis")

##plot no 2
x=np.array([3,6,7,9])
y= np.array([2,6,7,8])

plt.subplot(2,4,2)
plt.plot(x,y)
plt.title("plot no 2")
plt.xlabel("x axis")
plt.ylabel("y axis")


##plot no 3
x=np.array([1,4,5,9])
y= np.array([3,4,5,9])

plt.subplot(2,4,3)
plt.plot(x,y)
plt.title("plot no 2")
plt.xlabel("x axis")
plt.ylabel("y axis")


##plot no 4
x=np.array([5,3,2,1])
y=np.array([1,2,3,7])

plt.subplot(2,4,4)
plt.plot(x,y)
plt.title("plot no 4")
plt.xlabel("x axis")
plt.ylabel("y axis")



##plot no 5
x=np.array([1,2,3,4])
y= np.array([1,3,4,3])

plt.subplot(2,4,5)
plt.plot(x,y)
plt.title("plot no 5")
plt.xlabel("x axis")
plt.ylabel("y axis")



##plot no 6
x=np.array([2,4,5])
y=np.array([6,7,5])

plt.subplot(2,4,6)
plt.plot(x,y)
plt.title("plot no 6")
plt.xlabel("x axis")
plt.ylabel("y axis")



##plot no 7
x=np.array([2,5,8])
y=np.array([3,4,5])

plt.subplot(2,4,7)
plt.plot(x,y)
plt.title("plot no 7")
plt.xlabel("x axis")
plt.ylabel("y axis")



##plot no 8
x=np.array([7,4,2])
y=np.array([5,3,1])

plt.subplot(2,4,8)
plt.plot(x,y)
plt.title("plot no 8")
plt.xlabel("x axis")
plt.ylabel("y axis")
```
